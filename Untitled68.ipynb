```python
import os
from pycocotools.coco import COCO
import matplotlib.pyplot as plt
import cv2

# Path to the dataset
dataset_path = "/content/drive/MyDrive/insulators/train"
annotations_file = os.path.join(dataset_path, "_annotations.coco.json")

# Load the COCO dataset
coco = COCO(annotations_file)

# Display categories and their counts
categories = coco.loadCats(coco.getCatIds())
category_names = [cat['name'] for cat in categories]
print(category_names)
category_ids = coco.getCatIds()
category_counts = [len(coco.getAnnIds(catIds=cat_id)) for cat_id in category_ids]

# Plot dataset distribution
plt.figure(figsize=(2, 3))
print(category_counts[1:])
plt.bar(category_names, category_counts, color='skyblue')
plt.xlabel("Categories")
plt.ylabel("Number of Instances")
plt.title("Dataset Distribution by Category")
plt.show()

# Function to display an image with annotations
def show_image(image_id):
    img_info = coco.loadImgs(image_id)[0]
    img_path = os.path.join(dataset_path, img_info['file_name'])
    img = cv2.cvtColor(cv2.imread(img_path), cv2.COLOR_BGR2RGB)
    anns = coco.loadAnns(coco.getAnnIds(imgIds=image_id))

    plt.imshow(img)
    plt.axis("off")
    for ann in anns:
        x, y, w, h = ann['bbox']
        plt.gca().add_patch(plt.Rectangle((x, y), w, h, edgecolor='red', facecolor='none', lw=2))
        plt.text(x, y - 5, coco.loadCats([ann['category_id']])[0]['name'], color='red', fontsize=10, backgroundcolor='white')
    plt.show()

# # # Display first 5 images with annotations
# for img_id in coco.getImgIds()[16:24]:
#     show_image(img_id)
```

```python
import cv2
import torch
from pycocotools.coco import COCO
from torch.utils.data import Dataset, DataLoader
from torchvision.transforms import functional as F
from torchvision import transforms
from torchvision.models.detection import fasterrcnn_resnet50_fpn
from torch.optim import SGD
from torch.amp import GradScaler, autocast
from torch.optim.lr_scheduler import StepLR


import os
import yaml
from ultralytics import YOLO
from tqdm import tqdm
from pycocotools.coco import COCO
from PIL import Image
import matplotlib.pyplot as plt
import cv2
import torch

def prepare_yolo_dataset(coco_train_path, train_annotations_file, coco_val_path, val_annotations_file, coco_test_path, test_annotations_file):
    """
    Prepare dataset in YOLOv8 format directly using pre-split train, val, and test datasets without image resizing.
    """
    # Create directory structure
    os.makedirs('dataset/images/train', exist_ok=True)
    os.makedirs('dataset/images/val', exist_ok=True)
    os.makedirs('dataset/images/test', exist_ok=True)
    os.makedirs('dataset/labels/train', exist_ok=True)
    os.makedirs('dataset/labels/val', exist_ok=True)
    os.makedirs('dataset/labels/test', exist_ok=True)

    def convert_coco_to_yolo(coco_path, annotations_file, split):
        """
        Convert COCO annotations to YOLO format, mapping all classes to a single 'insulator' class.
        """
        # Load COCO dataset
        coco = COCO(annotations_file)

        # Define the target category mapping
        category_mapping = {}
        for cat in coco.loadCats(coco.getCatIds()):
            if cat['name'] in ['defect', 'insulator']:
                category_mapping[cat['id']] = 0  # Map to 'insulator' class ID (0)
            else:
                category_mapping[cat['id']] = None  # Exclude other classes

        # Initialize object count for classes
        object_count = {class_id: 0 for class_id in set(category_mapping.values()) if class_id is not None}

        # Process each image
        for img_id in tqdm(coco.getImgIds(), desc=f'Processing {split} set'):
            # Load image info
            img_info = coco.loadImgs(img_id)[0]
            img_path = os.path.join(coco_path, img_info['file_name'])

            # Get annotations
            ann_ids = coco.getAnnIds(imgIds=img_id)
            anns = coco.loadAnns(ann_ids)

            # Convert annotations to YOLO format
            with open(f'dataset/labels/{split}/{os.path.splitext(img_info["file_name"])[0]}.txt', 'w') as f:
                for ann in anns:
                    cat_id = category_mapping.get(ann['category_id'], None)
                    if cat_id is None:
                        continue  # Skip excluded categories

                    # Increment object count for the class
                    object_count[cat_id] += 1

                    # Convert bbox from [x, y, width, height] to [x_center, y_center, width, height] normalized
                    x, y, w, h = ann['bbox']
                    x_center = (x + w / 2) / img_info['width']
                    y_center = (y + h / 2) / img_info['height']
                    w /= img_info['width']
                    h /= img_info['height']

                    # Write to file using normalized values
                    f.write(f'{cat_id} {x_center} {y_center} {w} {h}\n')


    # Convert train, val, and test annotations
    convert_coco_to_yolo(coco_train_path, train_annotations_file, 'train')
    convert_coco_to_yolo(coco_val_path, val_annotations_file, 'val')
    convert_coco_to_yolo(coco_test_path, test_annotations_file, 'test')

    # Create dataset.yaml
    data_yaml = {
        'path': os.path.abspath('dataset'),
        'train': 'images/train',
        'val': 'images/val',
        'test': 'images/test',
        'nc': 1,  # Single class (insulator)
        'names': ['insulator']  # Define single class name
    }

    with open('dataset.yaml', 'w') as f:
        yaml.dump(data_yaml, f)

    return 'dataset.yaml'

def plot_test_predictions(model, test_image_dir, num_samples=5):
    """
    Plot predictions for a few test images.
    """
    test_images = os.listdir(test_image_dir)
    for i, img_file in enumerate(test_images[:num_samples]):
        img_path = os.path.join(test_image_dir, img_file)
        results = model.predict(source=img_path, save=False, conf=0.5)

        # Plot the image and predictions
        img = cv2.imread(img_path)
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        plt.figure(figsize=(10, 10))
        plt.imshow(img)

        for bbox in results[0].boxes.xyxy.cpu().numpy():
            x1, y1, x2, y2 = bbox[:4]
            plt.gca().add_patch(
                plt.Rectangle((x1, y1), x2 - x1, y2 - y1, fill=False, edgecolor='red', linewidth=2)
            )
        plt.title(f"Predictions for {img_file}")
        plt.axis('off')
        plt.show()

def train_yolo_model(data_yaml_path, epochs=2):
    """
    Train YOLOv8 model
    """
    # Load a model
    model = YOLO('yolov8n.pt')  # load a pretrained model (recommended for training)

    # Train the model
    results = model.train(
        data=data_yaml_path,
        epochs=epochs,
        imgsz=640,
        batch=16,
        device='0',
        save=True
    )

    # Retrieve the path of the best model
    best_model_path = model.ckpt_path if hasattr(model, 'ckpt_path') else None  # Ensure attribute exists

    # Define results directory explicitly if not available
    results_dir = './runs/detect/train'  # Default YOLO results directory

    # Print key metrics
    print(f"Training completed!\n"
          f"- Best Model Path: {best_model_path}\n"
          f"- Results saved to: {results_dir}")

    return model, best_model_path

def main():
    # Prepare dataset with resizing
    data_yaml_path = prepare_yolo_dataset(
        coco_train_path="/content/drive/MyDrive/insulators/train",
        train_annotations_file="/content/drive/MyDrive/insulators/train/_annotations.coco.json",
        coco_val_path="/content/drive/MyDrive/insulators/valid",
        val_annotations_file="/content/drive/MyDrive/insulators/valid/_annotations.coco.json",
        coco_test_path="/content/drive/MyDrive/insulators/test",
        test_annotations_file="/content/drive/MyDrive/insulators/test/_annotations.coco.json",
        # resize_to=(1024, 1024)  # Set desired image size here
    )

    # Train model
    model, _ = train_yolo_model(data_yaml_path, epochs=108)

    # Plot predictions on test data
    plot_test_predictions(model, 'dataset/images/test', num_samples=5)

if __name__ == "__main__":
    main()
```

```python
import os
from ultralytics import YOLO
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np


def test_model_with_metrics(model_path, data_yaml_path, confidence_threshold=0.5):
    """
    Test a YOLO model on the labeled test set, display metrics, and plot a confusion matrix.

    Args:
        model_path (str): Path to the trained YOLO model (.pt file).
        data_yaml_path (str): Path to the dataset YAML file.
        confidence_threshold (float): Confidence threshold for predictions.

    Returns:
        None
    """
    # Load the model
    print("Loading model...")
    model = YOLO(model_path)

    # Perform testing (using the `val` method for evaluation on the test set)
    print("Testing model...")
    metrics = model.val(data=data_yaml_path, split='test', conf=confidence_threshold)

    # Extract confusion matrix and other metrics
    confusion_matrix = metrics.confusion_matrix  # Access confusion matrix
    class_names = metrics.names  # Get class names

    # Plot confusion matrix
    plot_confusion_matrix(confusion_matrix, class_names)

    # Print key metrics
    print("\nKey Metrics:")
    print(f"mAP@50: {metrics.map50:.4f}")
    print(f"mAP@50-95: {metrics.map:.4f}")
    print(f"Precision: {metrics.precision:.4f}")
    print(f"Recall: {metrics.recall:.4f}")


def plot_confusion_matrix(conf_matrix, class_names):
    """
    Plot the confusion matrix.

    Args:
        conf_matrix (np.ndarray): Confusion matrix (2D array).
        class_names (list): List of class names.

    Returns:
        None
    """
    plt.figure(figsize=(10, 8))
    sns.heatmap(conf_matrix, annot=True, fmt="d", cmap="Blues", xticklabels=class_names, yticklabels=class_names)
    plt.xlabel("Predicted Labels")
    plt.ylabel("True Labels")
    plt.title("Confusion Matrix")
    plt.show()


if __name__ == "__main__":
    # Path to the trained YOLO model
    model_path = "/content/runs/detect/train4/weights/best.pt"

    # Path to the dataset YAML file
    data_yaml_path = "/content/dataset.yaml"

    # Confidence threshold for predictions
    confidence_threshold = 0.5

    # Test model and display metrics
    test_model_with_metrics(model_path, data_yaml_path, confidence_threshold)
```
